```python
# Run in Colab (after you already created `client = HUBClient(credentials)` as before)
import os, requests, shutil, sys

model_id = "yR8RJAYvVe1zjXlbzk24"   # your model id
out_dir = "/content/"               # change to a Drive path if you've mounted Drive, e.g. "/content/drive/MyDrive/ultralytics_models/"

# get the model object
model = client.model(model_id)

# 1) ask HUB for available weights URLs
print("Querying HUB for weights URLs...")
try:
    weights_info = model.get_weights_url()   # this usually returns dict like {'best': 'https://...', 'last': 'https://...'}
except Exception as e:
    print("Error calling get_weights_url():", e)
    print("Dumping model.data for inspection...")
    import json
    print(json.dumps(model.data, indent=2))
    raise

print("weights_info:", weights_info)

# Normalize to mapping name -> url
if isinstance(weights_info, str):
    mapping = {"weights": weights_info}
elif isinstance(weights_info, (list, tuple)):
    # list of urls or dicts
    mapping = {}
    for i, entry in enumerate(weights_info):
        if isinstance(entry, str):
            mapping[f"weights_{i}"] = entry
        elif isinstance(entry, dict):
            # merge dict entries
            for k,v in entry.items():
                mapping[str(k)] = v
        else:
            mapping[f"entry_{i}"] = str(entry)
elif isinstance(weights_info, dict):
    mapping = weights_info
else:
    mapping = {"weights": str(weights_info)}

if not mapping:
    raise RuntimeError("No weights URLs found in weights_info: {}".format(weights_info))

print("Resolved weight URLs:")
for k,v in mapping.items():
    print(k, "->", v)

# 2) Download each URL with the same headers the SDK uses (auth)
headers = {}
try:
    # model.headers exists on the object per your dir() output
    headers = model.headers or {}
except Exception:
    # try to get headers from hub_client
    try:
        headers = model.hub_client.headers
    except Exception:
        headers = {}

print("Using headers keys:", list(headers.keys()))

os.makedirs(out_dir, exist_ok=True)

downloaded = []
for name, url in mapping.items():
    # determine filename
    fname = os.path.basename(url.split("?")[0]) or f"{name}.pt"
    dest = os.path.join(out_dir, fname)
    print(f"\nDownloading {name} -> {fname} ...")
    try:
        with requests.get(url, headers=headers, stream=True, timeout=60) as r:
            r.raise_for_status()
            total = r.headers.get('content-length')
            total = int(total) if total is not None else None
            with open(dest, 'wb') as f:
                downloaded_bytes = 0
                for chunk in r.iter_content(chunk_size=8192):
                    if chunk:
                        f.write(chunk)
                        downloaded_bytes += len(chunk)
                print(f"Saved {downloaded_bytes} bytes to {dest}")
        downloaded.append(dest)
    except Exception as e:
        print(f"Failed to download {url}: {e}")
        # save the response content for debugging (if any)
        try:
            resp = requests.get(url, headers=headers, timeout=20)
            print("Second attempt status:", resp.status_code)
            print("Response headers:", resp.headers)
            print("Response text (first 400 chars):", resp.text[:400])
        except Exception as e2:
            print("Second attempt failed:", e2)

# 3) list files in out_dir
print("\nDownload complete. Files in", out_dir)
for line in os.listdir(out_dir):
    print(" -", line)

if not downloaded:
    raise RuntimeError("No files were downloaded. Check permissions / API key / model access.")
else:
    print("\nDownloaded files:")
    for p in downloaded:
        print(p)
```
